```python
import pandas as pd
```

```python
df = pd.read_csv('diabetic_data.csv')
```

```python
df['patient_nbr'] = df['patient_nbr'].astype(str)
df['encounter_id'] = df['encounter_id'].astype(str)
df['admission_type_id'] = df['admission_type_id'].astype(str)
df['discharge_disposition_id'] = df['discharge_disposition_id'].astype(str)
df['admission_source_id'] = df['admission_source_id'].astype(str)
```

```python
df.replace('?', pd.NA, inplace=True)
```

```python
df.isnull().sum()
```

```python
# df['max_glu_serum'].fillna('None', inplace=True)
# df['A1Cresult'].fillna('None', inplace = True)

df.fillna({'max_glu_serum':'0', 'A1Cresult': '0','race':'Other'}, inplace=True)
```

```python
print(df['weight'].value_counts())
print(df['payer_code'].value_counts())
print(df['medical_specialty'].value_counts())
```

```python
df.drop(['weight', 'payer_code', 'medical_specialty'], axis=1, inplace=True)
```

```python
df.drop(df[df.diag_1.isna()].index, inplace = True)

df.drop(df[df.diag_2.isna()].index, inplace = True)
df.drop(df[df.diag_3.isna()].index, inplace = True)
df.reset_index(inplace = True, drop = True)
```

```python
df.info()
```

```python
df.age.value_counts()
```

```python
# df['readmitted'].replace({'NO': 'NO', '>30': 'YES', '<30': 'YES'}, inplace=True)
df.replace({'readmitted': {'NO': 0, '>30': 1, '<30': 1}}, inplace=True)
```

```python
df.gender.value_counts()
```

```python
df = df[df.gender != 2.0]
```
